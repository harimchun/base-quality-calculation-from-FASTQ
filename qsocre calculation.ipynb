```python
a = "H?H;GHGHBHH>FHDE>@BE3>D>45:8>;.2<75@C7A#############################################################"
b = "@###################################################################################################"
```

```python
a_score = []
b_score = []
x_range = range(1, 101)
for i in a:
    a_score.append(ord(i)-33)
for i in b:
    b_score.append(ord(i)-33)
    
a_err_rate = []
b_err_rate = []
for i in a_score:
    a_err_rate.append(10**(-i/10))
for i in b_score:
    b_err_rate.append(10**(-i/10))
```

```python
print(a_score)
print(b_score)
```

```python
print(a_err_rate)
print(b_err_rate)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(x_range, a_score, label = "First Read")
plt.plot(x_range, b_score, label = "Scond Read")
plt.xlabel("Position in read(bp)")
plt.ylabel("score")
plt.title("Quality Scores across all bases")
plt.legend()
plt.savefig('fig1.png', dpi=300)
```

```python
plt.plot(x_range, a_err_rate, label = "First Read")
plt.plot(x_range, b_err_rate, label = "Scond Read")
plt.xlabel("Position in read(bp)")
plt.ylabel("Error Rate")
plt.title("Error Rate across all bases")
plt.legend()
plt.savefig('fig2.png', dpi=300)
```
